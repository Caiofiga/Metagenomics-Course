# Instalação da ferramenta FASTQC

Usando o terminal, faça a instalação da ferramenta FASTQC com a seguinte linha de código:  


```
$ mamba install -c bioconda fastqc
```



Agora faça um teste no terminal para verificar se a instalação ocorreu como esperado:  


```
$ fastqc --help
```



# Relatórios de qualidade com FASTQC

O [FASTQC](https://www.bioinformatics.babraham.ac.uk/projects/fastqc/) é uma ferramenta muito utilizada por bioinformatas para gerar relatórios de qualidade de sequenciamentos.  
Experimente trazer o arquivo .FASTQ para o servidor e executar no terminal:  


```
$ fastqc nome_arquivo.fastq
```



Agora faça download do relatório que foi gerado pelo FASTQC (arquivo .html) na janela de arquivos do lado esquerdo.
